## 3.1.2. Hypothesis testing: comparing two groups

### 3.1.2.1. Studentâs t-test: the simplest statistical test

#### 1-sample t-test: testing the value of a population mean

```python
stats.ttest_1samp(data['VIQ'], 0) 
```

#### 2-sample t-test: testing for difference across populations

```python
female_viq = data[data['Gender'] == 'Female']['VIQ']
male_viq = data[data['Gender'] == 'Male']['VIQ']
stats.ttest_ind(female_viq, male_viq)
```

### 3.1.2.2. Paired tests: repeated measurements on the same individuals

### To test if FISQ and PIQ are significantly different. We can use a 2 sample test:

```python
stats.ttest_ind(data['FSIQ'], data['PIQ'])
```

```python
data = pandas.read_csv('brain_size.csv', sep=';', na_values='.')
plt.figure(figsize=(4, 3))
data.boxplot(column=['FSIQ', 'PIQ'])
```

#### The variance due to inter-subject variability is confounding, and can be removed, using a âpaired testâ, or ârepeated measures testâ:

```python
stats.ttest_rel(data['FSIQ'], data['PIQ'])  
```

#### This is equivalent to a 1-sample test on the difference:

```python
stats.ttest_1samp(data['FSIQ'] - data['PIQ'], 0) 
```

```python
plt.figure(figsize=(4, 3))
plt.boxplot(data['FSIQ'] - data['PIQ'])
plt.xticks((1, ), ('FSIQ - PIQ', ))
```

## 3.1.3.2. Multiple Regression: including multiple factors

##### A linear model explaining a variable z (the dependent variable) can be seen in 3D as fitting a plane to a cloud of (x, y, z) points.

```python
data = pandas.read_csv('iris.csv')
print(data)
```

```python
model = ols('sepal_width ~ name + petal_length', data).fit()
print(model.summary()) 
```

```python
cluster_colors = {'setosa': 'purple', 'versicolor': 'green', 'virginica': 'yellow'}
colors = [cluster_colors[label] for label in data['name']]
scatter_matrix_plot =plotting.scatter_matrix(data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']],color=colors)
title_text = 'Blue: Setosa, Green: Versicolor, Red: Virginica'
plt.suptitle(title_text, fontsize=11)
```

### 3.1.3.3. Post-hoc hypothesis testing: analysis of variance (ANOVA)

#### To test if the petal length is different between versicolor and virginica, after removing the effect of sepal width, we want to test "name[T.versicolor] - name[T.virginica]", with an F-test:

```python
print(model.f_test([0, 1, -1, 0])) 
```

##### Since, the p value is greater than 0.05, null hypothesis is not rejected. Therefore, there is no significance difference in the petal length between versicolor and virginica.

### Exercise 5

```python
#### Going back to the brain size + IQ data, test if the VIQ of male and female are different after removing the effect of brain size, height and weight.

data = pandas.read_csv('brain_size.csv', sep=';', na_values=".")

df = data[['VIQ', 'Gender', 'MRI_Count', 'Height', 'Weight']]
df = df.dropna()
df_encoded = pandas.get_dummies(df, columns=['Gender'], drop_first=True)
model = sm.OLS(df_encoded['VIQ'], sm.add_constant(df_encoded[['MRI_Count', 'Height', 'Weight']]))
result = model.fit()

ssr = result.ssr
df_regression = result.df_model
msr = ssr / df_regression

sse = result.ssr + result.centered_tss
df_residual = result.df_resid
mse = sse / df_residual

f_statistic = msr / mse
p_value = 1 - stats.f.cdf(f_statistic, df_regression, df_residual)

anova_table = pandas.DataFrame({'df_regression': [df_regression],
                            'ssr': [ssr],
                            'msr': [msr],
                            'df_residual': [df_residual],
                            'sse': [sse],
                            'mse': [mse],
                            'f_statistic': [f_statistic],
                            'p_value': [p_value]})

anova_table.index = ['ANOVA']
anova_table.index.name = 'Source'

print(anova_table)

##### Since, the p value is greater than 0.05, null hypothesis is not rejected. Therefore, there is no significance difference in the VIQ of male and female
```
